```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

#funded companies 
r = requests.get('http://www.seed-db.com/companies/funding?value=1000000')
soup = BeautifulSoup(r.text)

table = soup.find('tbody')
table_data = []
for row in table.findAll('tr'):
   row_data = []
   for cell in row.findAll('td'):
       cell_data = filter(lambda x: x, map(lambda x: x.strip(), (list(cell.strings)))) #strip extra space
       row_data.append(' '.join(cell_data))
   table_data.append(row_data)
   
df = pd.DataFrame(table_data)
```
