```python
var_to_drop = ['var1','var2','var3','var5', 'var6', 'var12', 'var14', 'var16']
xtrain.drop(var_to_drop, axis =1)
xtest.drop(var_to_drop, axis =1)
#drop all the var1-17 columns that have more than 50% of NA. 
```

```python
def make_dummies(df, variables):
    for variable in variables:
        dummies = pd.get_dummies(df[variable], prefix = variable)
        df = pd.concat([df, dummies], axis = 1)
        df = df.drop(variable, axis = 1)
    return df

xtrain = make_dummies(xtrain, ['var4', 'var7','var8','var9'])
xtest = make_dummies(xtest, ['var4', 'var7','var8','var9'])
#categorical variables need to make 1,0 dummies
```

```python
xtrain['var15'] = xtrain['var15'] .fillna(xtrain['var15'].mean())
xtest['var15'] = xtest['var15'] .fillna(xtest['var15'].mean())
#var15 has 21.8% of NaN, fill in the missing with mean.
```

```python
df_target = pd.DataFrame(df_train[['target','target_binary','logloss']].values)
df_target.columns = ['target','target_binary','logloss']
```

```python
xtrain_final = pd.concat([xtrain, df_target],axis = 1)
```

```python
xtrain_final.to_csv('xtrain_final.csv', sep=',')
xtest.to_csv('xtest_final.csv', sep=',')
```
